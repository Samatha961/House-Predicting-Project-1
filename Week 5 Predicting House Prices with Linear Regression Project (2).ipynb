```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('house_data.csv')
```

```python
X = data.iloc[:, :-1].values  # All columns except the last one as features
y = data.iloc[:, -1].values   # Last column as the target (house price)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Step 6: Evaluate Model Performance using MSE and RÂ²
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"RÂ² Score: {r2:.2f}")
```

```python

# Optional: Display Actual vs Predicted Values
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison.head())
```
